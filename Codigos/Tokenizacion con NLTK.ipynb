```python
import nltk
import pandas as pd 
import numpy as np 
import re
```

```python
nltk.download('cess_esp')
```

```python
corpus = nltk.corpus.cess_esp.sents()
print(corpus)
print(len(corpus))
```

```python
flatten = [w for l in corpus for w in l]
```

```python
print(flatten[500:520])
```

```python
arr = [w for w in flatten if re.search('es',w)]
print(arr)
```

```python
arr = [w for w in flatten if re.search('es$',w )] #el $ indica que la palabra debe terminar en es
print(arr[:5])
```

```python
arr = [w for w in flatten if re.search('^es',w )] #el ^ indica que la palabra debe empezar en es
print(arr[:5])
```

```python
# Rango [a-z] o que tenga alguna [ghi]
arr = [w for w in flatten if re.search('^[ghi]',w)] #palabras que empiezan por g,h o i
print(arr[:10])
```

```python
# Clausura
# * repetir 0 o más veces
# + repetir 1 o mas veces
arr = [w for w in flatten if re.search('^(no)*',w)] #aparece no 0 o más veces 
print(arr[:10])

arr = [w for w in flatten if re.search('^(no)+',w)] #aparece no 1 o más veces 
print(arr[:10])
```

```python
texto = """ Cuando sea el rey del mundo (imaginaba él en su cabeza) no tendré que preocuparme
 por estas bobadas. Era solo un niño de 7 años, pero pensaba que podría ser cualquier 
 cosa que su imaginación le permitiera visualizar en su cabeza"""
```

```python
# Caso1: Tokenizar por espacios vacios usando libreria Re
print(re.split(r' ',texto))
```

```python
#Caso 2 Tokenización usando expresiones regulares
print(re.split(r'[ \t\n]+',texto))
```

```python
#Caso 3
print(re.split(r'[ \W\t\n]+',texto))
```

```python
#Caso 4 
texto = 'En los E.U. esa postal vale $15.50 ...'
print(re.split(r'[ \W\t\n]+',texto))
```

```python
pattern = r'''(?x)                 # Flag para iniciar el modo verbose
              (?:[A-Z]\.)+         # Hace match con abreviaciones como U.S.A.
              | \w+(?:-\w+)*       # Hace match con palabras que pueden tener un guión interno
              | \$?\d+(?:\.\d+)?%? # Hace match con dinero o porcentajes como $15.5 o 100%
              | \.\.\.             # Hace match con puntos suspensivos
              | [][.,;"'?():-_`]   # Hace match con signos de puntuación
'''
nltk.regexp_tokenize(texto,pattern)
```
